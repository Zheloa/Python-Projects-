## Task #1

```python
## Clear the duplicates
applications_df = applications_df.drop_duplicates(subset = 'applicant_id')
applications_df
```

```python
## Fill NA values with 0 in the 'External Rating' column
applications_df['External Rating'].fillna(0, inplace=True)
applications_df
```

```python
## Fill NA values with 'Середня' in the 'Education level' column
applications_df['Education level'].fillna('Середня', inplace=True)
applications_df
```

## Task #2 

```python
## Merging applications dataframe with industries dataframe
merged_df = applications_df.merge(industries_df, on = 'Industry')
merged_df
```

## Task #3

```python
## Creating a new column
merged_df['Rating'] = 0
## adding ratings using different conditions
merged_df['Rating'] = ((merged_df['Age'] >= 35) & (merged_df['Age'] <= 55)) * 20 + (merged_df['Marital status'] == 'Married') * 20 \
+ (merged_df['Location'].str.contains('Київ', na = False)) * 10 + (merged_df['Score']) + (merged_df['External Rating'] >= 7) * 20 \
+ (merged_df['External Rating'] <= 2) * (-20) 
merged_df
```

```python
## +20 rating for those who applied from Monday to Friday
merged_df['Applied at'] = pd.to_datetime(merged_df['Applied at'], format = 'mixed')
def add_points(row):
    if row['Applied at'].weekday() not in [5, 6]:
        return row['Rating'] + 20
    else:
        return row['Rating']

merged_df['Rating'] = merged_df.apply(add_points, axis=1)
merged_df
```

```python
## Limit the rating from 0 to 100
merged_df['Rating'].clip(lower=0, upper=100)
## If amount is null value or External rating = 0, then Rating = 0
merged_df.loc[merged_df['Amount'].isna() | (merged_df['External Rating'] == 0), 'Rating'] = 0
merged_df

```

## Task #4

```python
## Left only relevant ratings with value more than 0
merged_df = merged_df[merged_df['Rating'] > 0]
merged_df
```

## Task #5

```python
## Created a new column to find # of weeks
merged_df['Week'] = merged_df['Applied at'].dt.isocalendar().week
merged_df
```

```python
## Created the pivot table to find the average rating per week
table = merged_df.pivot_table(values = 'Rating', index = 'Week', aggfunc = 'mean')
table
```
